```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ershou = pd.read_csv("pachong/ershou.csv")
```

```python
ershou["size"].dtype
```

```python
ershou.info()
```

```python
ershou.describe()
```

```python
def zhuan(x):
    try:
        np.float64(x)
    except ValueError as e:
        x = np.float64(0.0)
    return x    
```

```python
ershou["size"] = np.float64(ershou["size"].map(zhuan))
```

```python
ershou["size"].dtype
```

```python
ershou.info()
```

```python
df = ershou.copy()
df['PerPrice'] = ershou['price']/ershou['size']
```

```python
df
```

```python
df_count = df.groupby('region')['price'].count().sort_values(ascending=False).to_frame().reset_index()
```

```python
df_mean = df.groupby('region')['PerPrice'].mean().sort_values(ascending=False).to_frame().reset_index()
```

```python
import seaborn as sns

plt.style.use("fivethirtyeight")
sns.set_style({'font.sans-serif':['simhei','Arial']})
```

```python
f, [ax1,ax2,ax3] = plt.subplots(3,1,figsize=(20,15))
sns.barplot(x='region', y='PerPrice', palette="Blues_d", data=df_mean, ax=ax1)

ax1.set_title('北京各大区二手房每平米单价对比',fontsize=15)
ax1.set_xlabel('区域')
ax1.set_ylabel('每平米单价')

sns.barplot(x='region', y='price', palette="Greens_d", data=df_count, ax=ax2)
ax2.set_title('北京各大区二手房数量对比',fontsize=15)
ax2.set_xlabel('区域')
ax2.set_ylabel('数量')
 
sns.boxplot(x='region', y='price', data=df, ax=ax3)
ax3.set_title('北京各大区二手房房屋总价',fontsize=15)
ax3.set_xlabel('区域')
ax3.set_ylabel('房屋总价')
 
plt.show()
```

```python

f, [ax1,ax2] = plt.subplots(1, 2, figsize=(15, 5))
# 建房时间的分布情况
sns.distplot(df['size'], bins=20, ax=ax1, color='r')
sns.kdeplot(df['size'], shade=True, ax=ax1)
# 建房时间和出售价格的关系
sns.regplot(x='size', y='price', data=df, ax=ax2)
plt.show()
```

```python
df['renovation'].value_counts()
```

```python
df['renovation'] = df.loc[(df['renovation'] != '南北'), 'renovation']

f, [ax1,ax2,ax3] = plt.subplots(1, 3, figsize=(20, 5))
sns.countplot(df['renovation'], ax=ax1)
sns.barplot(x='renovation', y='price', data=df, ax=ax2)
sns.boxplot(x='renovation', y='price', data=df, ax=ax3)
plt.show()
```
